```python
import numpy as np
```

```python
carretera = np.linspace(0, 11.5, num=1000)
```

```python
dist_media = []
for i in carretera:
    dist_media.append(sum(abs(c - i * np.ones(10)))/10)
```

```python
dist_media
```

```python
import matplotlib.pyplot as plt

plt.plot(carretera, dist_media, 'r-')
plt.show()
```

```python
carretera[np.argmin(dist_media)]
```
